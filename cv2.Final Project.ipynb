```python
import cv2
import numpy as np
```

```python
#Video Capturing 
# naming the new window
cap=cv2.VideoCapture(0)
cv2.namedWindow("sketch")

#Creating a while loop 
# reading the frame from cap. then make it gray scale and blur it.
#Apply canny for edges. Then apply threshold to convert it into defined binary digits.
#if in threshold we select Thresh_BINARY_INV the background can be white

while True:
    
    ret,frame=cap.read()
    image=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    image=cv2.GaussianBlur(image,(7,7),0)
    
    image=cv2.Canny(image,10,60)
    ret,image=cv2.threshold(image,50,255,cv2.THRESH_BINARY_INV)
    cv2.imshow("original",frame)
    cv2.imshow("sketch",image)
    if cv2.waitKey(1)==ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#How can we manually put trackbars for Canny during the live video?
#first we define a function

def nothing(x):
    pass


#Video Capturing 
# naming the new window
cap=cv2.VideoCapture(0)
cv2.namedWindow("sketch")

#We have to create a trackbar with LOW track Canny and UP Track Canny

cv2.createTrackbar("LTC","sketch",0,255,nothing)
cv2.createTrackbar("UTC","sketch",0,255,nothing)



#Creating a while loop 
# reading the frame from cap. then make it gray scale and blur it.
#Apply canny for edges. Then apply threshold to convert it into defined binary digits.
#if in threshold we select Thresh_BINARY_INV the background can be white

while True:
    
    ret,frame=cap.read()
    image=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    image=cv2.GaussianBlur(image,(7,7),0)

#we have to create a variable for LTC and UTC
    
    ltc=cv2.getTrackbarPos("LTC","sketch")
    utc=cv2.getTrackbarPos("UTC","sketch")

# in canny we put ltc and utc to change it manually inside the image
    
    image=cv2.Canny(image,ltc,utc)
    ret,image=cv2.threshold(image,50,255,cv2.THRESH_BINARY_INV)
    cv2.imshow("original",frame)
    cv2.imshow("sketch",image)
    if cv2.waitKey(1)==ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#comnination of sketch and live video 


cap=cv2.VideoCapture(0)

#We to select a region of interest in an image from webcom with the following command

while True:
    ret,im=cap.read()
    roi=cv2.selectROI("original",im,False,False)
    cv2.destroyAllWindows()
    break
    
# Then we have to use to video to first choose rectangle frame we want. then we have to convert  the rectanle to gray scale
#Roi[3] is the lengh, roi[2] is the breadth. Therefore, by adding them roi[1] and roi[0] we have our rectangle frame
# Then Gaussian Blurr and Canny to blur and find the edges and then apply binary threshold

while True:
    ret,frame=cap.read()
    image=frame[int(roi[1]):int(roi[1]+roi[3]),
               roi[0]:int(roi[0]+roi[2])]
    image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    image=cv2.GaussianBlur(image,(7,7),0)
    image=cv2.Canny(image,14,64)
    ret,image=cv2.threshold(image,50,255,cv2.THRESH_BINARY_INV)
    
# For showing the sketch and orignal video in one window, we have to overlapp the frame to the image. 
# in the first step we have to change from Gray scale to BGR
# Then we have to make it equivalent in this loop that the selected frame is inside the image. 
    
    image=cv2.cvtColor(image,cv2.COLOR_GRAY2BGR)
    
    frame[int(roi[1]):int(roi[1]+roi[3]),
               roi[0]:int(roi[0]+roi[2])]=image
    
    cv2.imshow("sketch",frame)
    if cv2.waitKey(1)==ord("q"):
        break
    
cap.release()
cv2.destroyAllWindows()
```
